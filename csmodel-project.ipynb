# I. **About the Dataset**

---


## **Brief Description**

This dataset provides essential information about recognized indicators of heart disease. Some of the key indicators in the dataset include obesity (indicated by a high BMI), not getting enough sleep, diabetic, etc.. Identifying the patterns of a higher-than-normal likelihood of getting a heart disease is necessary for people to take the necessary precautionary measures.


## **Description of the Data Collection Process**

- The dataset is obtained from the Centers for Disease Control and Prevention (CDC) to collect data on the health status of US residents through a cross-sectional major annual telephone survey conducted by the Behavioral Risk Factor Surveillance System (BRFSS). The survey is conducted using Random Digit Dialing (RDD), which involves contacting individuals through landline phones and mobile phones.

- The BRFSS survey comprises three sections: the core modules, optional modules, and state-added modules. The core section is standardized across all states and includes questions mandated by the CDC. Optional modules are chosen by individual states. The states have the flexibility to create their own unique modules referred to as state-added questions. [Source](https://www.hhs.nd.gov/public-health/behavioral-health/north-dakota-behavioral-risk-factor-surveillance-system/brfss)

- BRFSS employs iterative proportional fitting or “raking” as a data weighting technique. Raking incorporates data from cellular telephone surveys and the inclusion of additional demographic variables such as educational level, marital status, and housing status. These additional characteristics significantly improve the accuracy and comprehensiveness of the BRFSS sample, ensuring it better reflects the diverse socio-demographic composition of individual states. [Source](https://www.cdc.gov/brfss/annual_data/2022/pdf/Overview_2022-508.pdf)

- The use of RDD to contact individuals through landlines and mobile phones may introduce sampling bias. Not everyone has access to phones, thus excluding certain demographic groups such as those without phone access or those who do not answer their phones. This affects the representativeness of the sample which may possibly lead to skewed conclusions. Additionally, the survey introduces response bias as respondents may not always answer the questions truthfully, which may lead to incorrect conclusions regarding health conditions.
- Since the survey conducted is a cross-sectional survey, the dataset may not capture the changes in health conditions over time. The conclusions and insights that will be generated are limited to the year it was conducted.


## **Structure Description**

- Each row represents an observation from a participant in the survey performed by the CDC, with each column representing an aspect of their health and lifestyle that was asked during the survey, such as their weight, their number of sleep hours, and if they have asthma.
- There are 445,132 observations in the dataset.
- There are 40 variables in the dataset.
- The dataset does not have separate, different files that need to be combined

## 4. What percentage of individuals in different age groups engage in smoking, regardless of the quantity?


### Variables Involved

- Age

Write the code to select SmokerStatus and Age as a dataframe.


```python
# Get the dataframe with age and smokerstatus
age_smoker_df = heart_df[['Age', 'SmokerStatus']]

# Get the values of the age column
unsorted_age_values = heart_df['Age'].unique()

# change the categorical type to int
age_values = []
for i in range(len(unsorted_age_values)):
    age_values.append(int(unsorted_age_values[i]))

# sort the values
age_values.sort()

# Create arrays for each of the age values
temp_arr = []

# separate the dataframes for each of the smoker status using query
for age_value in age_values:
    # convert int to str to be used with .query()
    age_str = str(age_value)
    # perform query
    temp_df = age_smoker_df.query('Age == ' + age_str)     
    # store the df to the array
    temp_arr.append(temp_df)

# Data
groups = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49',
          '50-54', '55-59', '60-64', '65-69', '70-74', '80 above']
never_smoked_values = []
former_smoker_values = []
smokes_some_days_values = []
smokes_everyday_values = []

# get the percentages of the smoker classifications in each age category
for i in range(len(groups)):
    never_smoked_values.append(
        (temp_arr[i].value_counts(normalize=True) * 100).values[0])
    former_smoker_values.append(
        (temp_arr[i].value_counts(normalize=True) * 100).values[1])
    smokes_some_days_values.append(
        (temp_arr[i].value_counts(normalize=True) * 100).values[2])
    smokes_everyday_values.append(
        (temp_arr[i].value_counts(normalize=True) * 100).values[3])
```

## Visualization


```python
# Use Stacked bar chart - https://python-charts.com/part-whole/stacked-bar-chart-matplotlib/
fig, ax = plt.subplots()

# configure the bars/handles to represent data
ax.bar(groups, never_smoked_values, edgecolor="black",
       linewidth=2, color="lightseagreen", label="Never Smoked")
ax.bar(groups, former_smoker_values, bottom=never_smoked_values,
       edgecolor="black", linewidth=2, color="yellowgreen", label="Former Smoker")
ax.bar(groups, smokes_some_days_values, bottom=np.add(
    never_smoked_values, former_smoker_values), edgecolor="black", linewidth=2, color="burlywood", label="Seldomly Smokes")
ax.bar(groups, smokes_everyday_values, bottom=np.add(never_smoked_values,
       np.add(former_smoker_values, smokes_some_days_values)), edgecolor="black", linewidth=2, color="tomato", label="Smokes Everyday")

# get the handles and labels using get_legend_handles_labels()
handles, labels = ax.get_legend_handles_labels()
# reverse the order of the labels and handles
ax.legend(handles[::-1], labels[::-1], loc='lower right')

# Set the axis labels
ax.set_ylabel("Percentage of Smokers/Non-Smokers")
ax.set_xlabel("Age Groups")

# Label the specific bars with the corresponding data (percentage in this case)
for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() / 2 + bar.get_y(),
            round(bar.get_height()), ha='center',
            color='white', weight='bold', size=10)

# rotate the x-axis labels for readability
labels = ax.get_xticklabels()
for label in labels:
    label.set_rotation(90)

# show the plot
plt.show()
```

# 5. Are there differences in the prevalence of heart diseases between alcohol drinkers and non-alcohol drinkers in the dataset?

### Variable/s involved:

- HadHeartDisease
- AlcoholDrinkers


### Numerical

The main numerical summary will be the distribution of those with heart diseases among alcohol drinkers and non-alcohol drinkers.
In order to determine whether or not the test is statistically-significant, we must get the p-value.


```python
contingency_table = pd.crosstab(
    heart_df['AlcoholDrinkers'], heart_df['HadHeartDisease'])
chi2, p, _, _ = chi2_contingency(contingency_table)
print("P-value:", p)
```

## 7. Is there a relationship between a participant’s number of sleep hours and whether or not they have had heart disease?


### Variables Involved

- SleepHours
- HadHeartDisease


### Numerical


```python
# Pearson Correlation Coefficient

heart_df_copy = heart_df['HadHeartDisease'].map({'Yes': 1, 'No': 0})
sleep_hours_df = pd.DataFrame(
    {'SleepHours': heart_df['SleepHours'], 'HadHeartDisease': heart_df_copy})
correlation_coefficient_pearson = sleep_hours_df['SleepHours'].corr(
    sleep_hours_df['HadHeartDisease'])

print(f"Pearson Correlation Coefficient: {correlation_coefficient_pearson}")
```

```python
# Biserial Correlation Coefficient
from scipy import stats
import pandas as pd

heart_df_copy = heart_df['HadHeartDisease'].map({'Yes': 1, 'No': 0})
correlation_coefficient_biserial = stats.pointbiserialr(
    heart_df_copy, heart_df['SleepHours'])[0]

print(f"Biserial Correlation Coefficient: {correlation_coefficient_biserial}")
```

### Visualization


```python


plt.hist(heart_df[heart_df['HadHeartDisease'] == 'No']
         ['SleepHours'], alpha=0.5, label='No Heart Disease', color="lightseagreen", edgecolor="black", linewidth=2)
plt.hist(heart_df[heart_df['HadHeartDisease'] == 'Yes']
         ['SleepHours'], alpha=0.5, label='Had Heart Disease', color="yellowgreen", edgecolor="black", linewidth=2)
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency')
plt.title('Relationship between Sleep Hours and Heart Disease')
plt.legend(loc='upper right')

plt.show()
```
