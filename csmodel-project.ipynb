## **Brief Description**
This dataset provides essential information about recognized indicators of heart disease. Some of the key indicators in the dataset include obesity (indicated by a high BMI), not getting enough sleep, diabetic, etc.. Identifying the patterns of a higher-than-normal likelihood of getting a heart disease is necessary for people to take the necessary precautionary measures.

## **Description of the Data Collection Process**
- The dataset is obtained from the Centers for Disease Control and Prevention (CDC) to collect data on the health status of US residents through a cross-sectional major annual telephone survey conducted by the Behavioral Risk Factor Surveillance System (BRFSS). The survey is conducted using Random Digit Dialing (RDD), which involves contacting individuals through landline phones and mobile phones.  

- The BRFSS survey comprises three sections: the core modules, optional modules, and state-added modules. The core section is standardized across all states and includes questions mandated by the CDC. Optional modules are chosen by individual states. The states have the flexibility to create their own unique modules referred to as state-added questions. [Source](https://www.hhs.nd.gov/public-health/behavioral-health/north-dakota-behavioral-risk-factor-surveillance-system/brfss)

- BRFSS employs iterative proportional fitting or “raking” as a data weighting technique. Raking incorporates data from cellular telephone surveys and the inclusion of additional demographic variables such as educational level, marital status, and housing status. These additional characteristics significantly improve the accuracy and comprehensiveness of the BRFSS sample, ensuring it better reflects the diverse socio-demographic composition of individual states. [Source](https://www.cdc.gov/brfss/annual_data/2022/pdf/Overview_2022-508.pdf)

- The use of RDD to contact individuals through landlines and mobile phones may introduce sampling bias. Not everyone has access to phones, thus excluding certain demographic groups such as those without phone access or those who do not answer their phones. This affects the representativeness of the sample which may possibly lead to skewed conclusions. Additionally, the survey introduces response bias as respondents may not always answer the questions truthfully, which may lead to incorrect conclusions regarding health conditions.
- Since the survey conducted is a cross-sectional survey, the dataset may not capture the changes in health conditions over time. The conclusions and insights that will be generated are limited to the year it was conducted.

## **Structure Description**
- Each row represents an observation from a participant in the survey performed by the CDC, with each column representing an aspect of their health and lifestyle that was asked during the survey, such as their weight, their number of sleep hours, and if they have asthma.
- There are 445,132  observations in the dataset.
- There are 40 variables in the dataset.
- The dataset does not have separate, different files that need to be combined

## **Variable Discussion**
1. **State**: The state where the participant of the dataset lives.
2. **Sex**: The biological sex of the participant.
3. **GeneralHealth**: Self-reported overall health status of the participant.
4. **PhysicalHealthDays**: The number of days in the past month that the participant experienced physical health issues or discomfort.

7. **PhysicalActivities**: Whether or not the participant does physical activities.
8. **SleepHours**: The number of hours of sleep a participant gets on average per day.
9. **RemovedTeeth**: The count of the participant's teeth that have been removed or are missing.
10. **HadHeartAttack** (HadHeartDisease): As per the dataset expert's recommendation, this variable is about whether or not the participant has had a heart disease in the past, not necessarily just a heart attack.
11. **HadAngina**: Whether or not the participant has had an episode of Angina (chest pain or discomfort) in the past.
12. **HadStroke**: Whether or not the participant has had a stroke in the past.
13. **HadAsthma**: Whether or not the participant has been diagnosed with asthma.
14. **HadSkinCancer**: Whether or not the participant has had a diagnosis of skin cancer.
15. **HadCOPD**: Whether or not the participant has Chronic Obstructive Pulmonary Disease (COPD).
16. **HadDepressiveDisorder**: Whether or not the participant has been diagnosed with a depressive disorder.
17. **HadKidneyDisease**: Whether or not the participant has been diagnosed with kidney disease.
18. **HadArthritis**: Whether or not the participant has been diagnosed with arthritis.
19. **HadDiabetes**: Whether or not the participant has been diagnosed with diabetes.
20. **DeafOrHardOfHearing**: Whether or not the participant is deaf or hard of hearing.
21. **BlindOrVisionDifficulty**: Whether or not the participant is blind or has difficulty with vision.
22. **DifficultyConcentrating**: Whether or not the participant has difficulty concentrating or staying focused.
23. **DifficultyWalking**: Whether or not the participant has difficulty walking or mobility.
24. **DifficultyDressingBathing**: Whether or not the participant has difficulty dressing or bathing.
25. **DifficultyErrands**: Whether or not the participant has difficulty running daily errands or tasks.
26. **SmokerStatus**: The participant's current smoking status (never smoked, former smoker, current smoker).
27. **ECigaretteUsage**: The frequency and extent of e-cigarette usage.
28. **ChestScan**: Whether or not the participant has had a chest scan or imaging for a specific medical purpose.
29. **RaceEthnicityCategory**: The racial and ethnic background or category of the participant, and whether or not they are of Hispanic ethnicity.
30. **AgeCategory**: The age group to which the participant belongs.
31. **HeightInMeters**: The height of the participant in meters.
32. **WeightInKilograms**: The weight of the participant in kilograms.
33. **BMI**: The Body Mass Index (BMI) of the participant, calculated based on height and weight.
34. **AlcoholDrinkers**: Whether or not the participant consumes alcoholic beverages.
35. **HIVTesting**: Whether or not the participant has undergone HIV testing.
36. **FluVaxLast12**: Whether the participant received a flu vaccine within the last 12 months.
37. **PneumoVaxEver**: Whether the participant has ever received a pneumococcal vaccine.
38. **TetanusLast10Tdap**: Whether or not the participant has received a tetanus vaccine or Tdap (tetanus, diphtheria, and pertussis) combination vaccine.
39. **HighRiskLastYear**: Whether the participant was classified as "high risk" for certain health conditions within the last year.
40. **CovidPos**: Whether or not the participant has ever tested positive for COVID-19.

# IV. **Data Preprocessing**
***
For some of the variables, pre-processing was implemented in order to make the representation of the values easier to read and analyze.

"HadHeartAttack" variable will be changed to "HadHeartDisease" to be easily understood.


```python
# Change HadHeartAttack to HadHeartDisease
df.rename(columns = {'HadHeartAttack':'HadHeartDisease'}, inplace = True)
```

A separate dataframe is made which only includes the 8 aforementioned variables that will be utilized in the study.

```python
heart_df = df[["GeneralHealth", "AgeCategory", "HadHeartDisease", "BMI", "AlcoholDrinkers", "SmokerStatus", "SleepHours", "PhysicalActivities"]]
heart_df
```

For the BMI variable, binning was used to separate certain numerical values into BMI-specific labels.

```python
heart_df['BMI'] = pd.cut(heart_df.BMI, bins=[0, 18.4, 24.9, 29.9, 34.9, 39.9, 49.9, 100],
                        labels=['Underweight', 'Normal', 'Overweight', 'Obese', 'Severely Obese', 'Morbidly Obese', 'Super Obese'])
heart_df
```

For the AgeCategory variable, the age groups were converted to age estimates in order to allow the datatype conversion from category to integer.

```python
heart_df['AgeCategory'] = heart_df['AgeCategory'].str[4:6]
heart_df.rename(columns = {'AgeCategory':'Age'}, inplace = True)
```

The data type of the AgeCategory will be changed to integer.

```python
heart_df['Age'] = heart_df['Age'].astype(int)
heart_df
```
